```python
# This is the most important cell in EDA.
# We check if SalePrice is normally distributed.
# Linear regression works best when the target is bell-shaped.
 
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
 
# --- Plot 1 : Raw SalePrice ---
axes[0].hist(df['SalePrice'], bins=60, color='#7F77DD', edgecolor='white', linewidth=0.4)
axes[0].axvline(df['SalePrice'].mean(),   color='red',    linestyle='--', linewidth=1.5,
                label=f"Mean   : ${df['SalePrice'].mean():,.0f}")
axes[0].axvline(df['SalePrice'].median(), color='orange', linestyle='--', linewidth=1.5,
                label=f"Median : ${df['SalePrice'].median():,.0f}")
axes[0].set_title('SalePrice — Raw', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Sale Price ($)')
axes[0].set_ylabel('Number of Houses')
axes[0].legend()
```

```python
# np.log1p(x) = log(x + 1)  → safe even if x has zeros
log_price = np.log1p(df['SalePrice'])
 
axes[1].hist(log_price, bins=60, color='#1D9E75', edgecolor='white', linewidth=0.4)
axes[1].axvline(log_price.mean(),   color='red',    linestyle='--', linewidth=1.5,
                label=f"Mean   : {log_price.mean():.2f}")
axes[1].axvline(log_price.median(), color='orange', linestyle='--', linewidth=1.5,
                label=f"Median : {log_price.median():.2f}")
axes[1].set_title('log(SalePrice) — After Transformation', fontsize=13, fontweight='bold')
axes[1].set_xlabel('log(Sale Price)')
axes[1].set_ylabel('Number of Houses')
axes[1].legend()
 
plt.suptitle('Target Variable Analysis', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/figures/01_saleprice_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
 
# --- Key stats ---
print(f"Skewness — Raw SalePrice         : {df['SalePrice'].skew():.4f}")
print(f"Skewness — log(SalePrice)        : {log_price.skew():.4f}")
print(f"\nMin  : ${df['SalePrice'].min():,}")
print(f"Max  : ${df['SalePrice'].max():,}")
print(f"Mean : ${df['SalePrice'].mean():,.0f}")
print(f"\nConclusion: SalePrice is right-skewed (skew > 1).")
print("We will apply np.log1p() before modeling in Phase 3.")
print("To reverse predictions later: np.expm1(prediction)")
```

```python
# Count how many nulls each column has
missing_count = df.isnull().sum()
 
# Keep only columns that actually have missing values
missing_count = missing_count[missing_count > 0].sort_values(ascending=False)
 
# Turn into percentage
missing_pct = (missing_count / len(df) * 100).round(1)
 
# Combine into a table
missing_df = pd.DataFrame({
    'Missing Count' : missing_count,
    'Missing %'     : missing_pct
})
print(f"Columns with missing values : {len(missing_df)}\n")
print(missing_df.to_string())
 
# --- Plot ---
plt.figure(figsize=(10, 8))
 
# colour by severity : red > 50%, orange 20-50%, purple < 20%
bar_colors = []
for pct in missing_pct.values:
    if pct > 50:
        bar_colors.append('#D85A30')   # red
    elif pct > 20:
        bar_colors.append('#EF9F27')   # orange
    else:
        bar_colors.append('#AFA9EC')   # purple
 
plt.barh(missing_count.index, missing_pct.values, color=bar_colors)
plt.axvline(50, color='red',    linestyle='--', alpha=0.4, label='50%')
plt.axvline(20, color='orange', linestyle='--', alpha=0.4, label='20%')
plt.xlabel('Missing (%)')
plt.title('Missing Values per Column', fontsize=13, fontweight='bold')
plt.gca().invert_yaxis()   # highest at the top
plt.legend()
plt.tight_layout()
plt.savefig('../reports/figures/02_missing_values.png', dpi=150, bbox_inches='tight')
plt.show()
 
# --- Classify each missing column ---
print("\nStructural absence (house doesn't HAVE it → fill with 'None'):")
structural = ['Pool QC', 'Misc Feature', 'Alley', 'Fence', 'Fireplace Qu',
              'Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond',
              'Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
print(structural)
 
print("\nGenuinely unknown data (impute with median or mode in Phase 3):")
genuine = ['Lot Frontage', 'Garage Yr Blt', 'Mas Vnr Area', 'Mas Vnr Type', 'Electrical']
print(genuine)
```

```python
#We look at each numerical feature one by one.
# Goal: spot skewness and unusual shapes.
 
num_features = [
    'Gr Liv Area', 'Lot Area', 'Total Bsmt SF',
    '1st Flr SF',  'Garage Area', 'Mas Vnr Area',
    'TotRms AbvGrd', 'Year Built'
]
 
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
axes = axes.flatten()   # turn 2D array into 1D so we can loop easily
 
for i, col in enumerate(num_features):
    data = df[col].dropna()   # ignore NaN values for plotting
    axes[i].hist(data, bins=40, color='#7F77DD', edgecolor='white', linewidth=0.3)
    axes[i].set_title(col, fontsize=11, fontweight='bold')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Count')
 
    # print skewness value on the plot
    skew = data.skew()
    color = 'red' if abs(skew) > 0.75 else 'green'
    axes[i].text(0.97, 0.95, f'skew: {skew:.2f}',
                 transform=axes[i].transAxes,
                 ha='right', va='top',
                 fontsize=9, color=color,
                 bbox=dict(boxstyle='round,pad=0.2', fc='white', alpha=0.7))
 
plt.suptitle('Numerical Feature Distributions', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/figures/03_numerical_distributions.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# See how many houses fall into each category.
 
cat_features = ['Neighborhood', 'MS Zoning', 'House Style', 'Bldg Type', 'Sale Type']
 
for col in cat_features:
    value_counts = df[col].value_counts()
 
    plt.figure(figsize=(12, 4))
    plt.bar(value_counts.index, value_counts.values,
            color='#5DCAA5', edgecolor='white', linewidth=0.5)
    plt.title(f'{col} — How many houses in each category?',
              fontsize=12, fontweight='bold')
    plt.xlabel(col)
    plt.ylabel('Number of Houses')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(f'../reports/figures/04_cat_{col.replace(" ", "_")}.png',
                dpi=150, bbox_inches='tight')
    plt.show()
 
    # flag rare categories (fewer than 10 houses)
    rare = value_counts[value_counts < 10]
    if len(rare) > 0:
        print(f"\n⚠  Rare categories in '{col}' (< 10 houses) — may cause issues in train/test split:")
        print(rare.to_string())
        print()
 
 
```

```python
#Pearson correlation tells us how strongly each numerical
# feature moves together with SalePrice.
# Range: -1 (opposite) to +1 (same direction)
 
all_num_with_target = df.select_dtypes(include=['int64', 'float64'])
corr_with_target = (all_num_with_target
                    .corr()['SalePrice']
                    .drop('SalePrice')
                    .sort_values(ascending=False))
 
print("Top 15 features correlated WITH SalePrice:")
print(corr_with_target.head(15).round(3))
print("\nBottom 5 (almost no relationship with SalePrice):")
print(corr_with_target.tail(5).round(3))
 
# --- Bar chart of top correlations ---
top20 = corr_with_target.head(20)
 
plt.figure(figsize=(10, 8))
bar_colors = ['#1D9E75' if c > 0 else '#D85A30' for c in top20.values]
plt.barh(top20.index, top20.values, color=bar_colors)
plt.axvline(0, color='black', linewidth=0.8)
plt.xlabel('Pearson Correlation with SalePrice')
plt.title('Top 20 Features Correlated with SalePrice', fontsize=13, fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('../reports/figures/05_feature_correlations.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
#Shows how ALL top features relate to each other.
# Look for dark red squares (strong positive correlation between features)
# → those pairs have multicollinearity — important for Phase 5 & 8.
 
top_features = corr_with_target.head(12).index.tolist() + ['SalePrice']
corr_matrix  = df[top_features].corr()
 
# mask = hide the upper triangle (it's a mirror of the lower)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
 
plt.figure(figsize=(13, 10))
sns.heatmap(
    corr_matrix,
    mask    = mask,
    annot   = True,        # show numbers inside each box
    fmt     = '.2f',       # 2 decimal places
    cmap    = 'RdYlGn',   # red = negative, green = positive
    center  = 0,
    linewidths  = 0.5,
    linecolor   = 'white',
    cbar_kws    = {'shrink': 0.8}
)
plt.title('Correlation Heatmap — Top Features + SalePrice',
          fontsize=13, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../reports/figures/06_correlation_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
 
print("Pairs to watch for multicollinearity (corr > 0.8):")
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        val = corr_matrix.iloc[i, j]
        if abs(val) > 0.80 and corr_matrix.columns[i] != 'SalePrice' and corr_matrix.columns[j] != 'SalePrice':
            print(f"  {corr_matrix.columns[i]}  ↔  {corr_matrix.columns[j]}  :  {val:.2f}")
```

```python
#Each scatter plot shows one feature vs SalePrice.
# A clear upward line = strong relationship = good feature.
# A random cloud = weak relationship = might not help the model.
scatter_features = [
    'Gr Liv Area', 'Total Bsmt SF', 'Year Built',
    'Garage Area', 'Mas Vnr Area', 'TotRms AbvGrd'
]
 
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()
 
for i, col in enumerate(scatter_features):
    clean = df[[col, 'SalePrice']].dropna()
 
    # scatter dots
    axes[i].scatter(clean[col], clean['SalePrice'],
                    alpha=0.3, color='#7F77DD', s=12, edgecolor='none')
 
    # trend line
    slope, intercept, r, p, se = stats.linregress(clean[col], clean['SalePrice'])
    x_line = np.linspace(clean[col].min(), clean[col].max(), 200)
    axes[i].plot(x_line, slope * x_line + intercept,
                 color='red', linewidth=1.5, linestyle='--')
 
    corr_val = clean[col].corr(clean['SalePrice'])
    axes[i].set_title(f'{col}\nr = {corr_val:.3f}', fontsize=11, fontweight='bold')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('SalePrice ($)')
 
plt.suptitle('Numerical Features vs SalePrice (with trend line)',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../reports/figures/07_scatter_vs_target.png', dpi=150, bbox_inches='tight')
plt.show()
 
```

```python
# A box plot shows the price RANGE for each category.
# If the boxes are at very different heights → that feature matters.
# If all boxes are at the same height → feature has no effect on price.
 
box_features = ['Overall Qual', 'Neighborhood', 'Kitchen Qual', 'Exter Qual']
 
for col in box_features:
    plt.figure(figsize=(14, 5))
 
    # sort categories by median SalePrice so the plot is easy to read
    order = (df.groupby(col)['SalePrice']
               .median()
               .sort_values()
               .index)
 
    sns.boxplot(
        data    = df,
        x       = col,
        y       = 'SalePrice',
        order   = order,
        palette = 'viridis',
        linewidth   = 0.8,
        fliersize   = 2      # size of outlier dots
    )
    plt.title(f'{col} vs SalePrice\n(sorted by median price)',
              fontsize=12, fontweight='bold')
    plt.xlabel(col)
    plt.ylabel('Sale Price ($)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.savefig(f'../reports/figures/08_box_{col.replace(" ", "_")}.png',
                dpi=150, bbox_inches='tight')
    plt.show()
 
```

```python
# Specific outlier found in the literature:
# Two houses with HUGE living area but suspiciously LOW price.
# These are likely partial sales — they will hurt our model.
 
plt.figure(figsize=(11, 6))
 
# all houses as small dots
plt.scatter(df['Gr Liv Area'], df['SalePrice'],
            alpha=0.35, color='#7F77DD', s=15, edgecolor='none',
            label='All houses')
 
# find the suspicious ones
outliers = df[(df['Gr Liv Area'] > 4000) & (df['SalePrice'] < 300000)]
 
# highlight them in red
plt.scatter(outliers['Gr Liv Area'], outliers['SalePrice'],
            color='red', s=100, zorder=5,
            edgecolor='black', linewidth=0.8,
            label=f'Outliers ({len(outliers)} houses)')
 
# label each one with its details
for _, row in outliers.iterrows():
    plt.annotate(
        f" #{int(row['Order'])}\n {int(row['Gr Liv Area'])} sqft\n ${int(row['SalePrice']):,}",
        xy      = (row['Gr Liv Area'], row['SalePrice']),
        fontsize = 8.5,
        color    = 'red'
    )
 
plt.xlabel('Above-Ground Living Area (sqft)')
plt.ylabel('Sale Price ($)')
plt.title('Outlier Detection — Large Area, Low Price',
          fontsize=13, fontweight='bold')
plt.legend()
plt.tight_layout()
plt.savefig('../reports/figures/09_outliers.png', dpi=150, bbox_inches='tight')
plt.show()
 
print("Outlier details:")
cols_to_show = ['Order', 'Gr Liv Area', 'SalePrice', 'Neighborhood', 'Sale Condition']
print(outliers[cols_to_show].to_string(index=False))
print("\nNote: 'Sale Condition' will likely say 'Partial' — confirms these are special transactions.")
print("Decision: remove these 2 houses in Phase 3.")
 
 
```
